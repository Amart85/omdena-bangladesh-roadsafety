```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
from time import time
def make_soup(url):
    content = requests.get(url)
    html_content=content.content
    soup=BeautifulSoup(html_content,'lxml')
    
    return soup
```

```python
data_accident=[]
accident_time=[]
links=[]
for i in range(1,49):
    print(i)
    url2="https://www.dhakatribune.com/hashtag/road-accident/page/"+str(i)
    for el in make_soup(url2).find_all("h4",class_='news-title'):
        data_accident.append(el.get_text())
    
    for data_time in make_soup(url2).find_all("div",class_='listing-time'):
        accident_time.append(data_time.get_text())
    
    for link in make_soup(url2).find_all('div',class_='top-news'):
        links.append(link.a['href'])

print(len(data_accident))
print(len(accident_time))
print(len(links))
```
