#### Adding **full text** from the Newspapare as a new column temporarily for convenience

```python
import pandas as pd
df = pd.read_csv('en_2.csv')
```

```python
df.head()
```

```python
import requests
from bs4 import BeautifulSoup

links = df.link

full_text = []

for link in links:

        URL = link
        page = requests.get(URL)
        soup = BeautifulSoup(page.content, 'html.parser')
        parahraphs = soup.find_all('p')
        para = []

        for i in list(range(len(parahraphs))):
            para.append(soup.find_all('p')[i].get_text())
            s = ' '.join(para)
        full_text.append(s)
```

```python
df['full_text'] = full_text
display(df.head(3))
df.to_csv('prothom_alo_1jan2019-31dec2020_v1.csv', index = False)
```
